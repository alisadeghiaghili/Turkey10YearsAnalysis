```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
def extractData(path):
    file = pd.read_csv(path, sep = ',', skiprows = 4)
    file.drop(labels = ['Country Code', 'Indicator Name', 'Indicator Code'], axis = 'columns', inplace = True)
    file = pd.melt(file, id_vars = 'Country Name', var_name = 'Year', value_name = 'GDPGrowth')
    file.Year = pd.to_numeric(file.Year, errors='coerce').fillna(0).astype(np.int64)
    file = file[(file.Year >= 2009) & (file.Year <= 2019)]
    file.reset_index(drop = True, inplace = True)
    return file
```

```python
US = extractData(path = r'https://raw.githubusercontent.com/alisadeghiaghili/Turkey10YearsAnalysis/main/USGDP.csv')
US
```

```python
France = extractData(path = r'https://raw.githubusercontent.com/alisadeghiaghili/Turkey10YearsAnalysis/main/FranceGDP.csv')
France
```

```python
UK = extractData(path = r'https://raw.githubusercontent.com/alisadeghiaghili/Turkey10YearsAnalysis/main/UKGDP.csv')
UK
```

```python
Italy = extractData(path = r'https://raw.githubusercontent.com/alisadeghiaghili/Turkey10YearsAnalysis/main/ItalyGDP.csv')
Italy
```

```python
Turkey = extractData(path = r'https://raw.githubusercontent.com/alisadeghiaghili/Turkey10YearsAnalysis/main/TurkeyGDP.csv')
Turkey["Country Name"] = "Turkey"
Turkey
```

```python
data = pd.concat([US, France, UK, Italy, Turkey])
data.reset_index(drop = True, inplace = True)
data
```

```python
import seaborn as sns
fig = plt.figure()
sns.pointplot(x = data.Year, 
              y = 'GDPGrowth', 
              hue ="Country Name", 
              data = data, 
              palette = "husl", 
              scale = 0.7, 
              linewidth = '')

sns.set_style("white", {'font-family': [u'Arial']})
sns.set_context('notebook', font_scale=.90)

sns.despine(left = True)
plt.xlabel('Year')
plt.ylabel('GDP Growth (%)')
plt.title('NATO Top 5 military powers GDP growth\n(10 Year)')

plt.gca().spines['bottom'].set_color('#e1e1e1')
plt.gca().spines['bottom'].set_linestyle('--')
fig.autofmt_xdate()
plt.gca().set_ylim(-10,15)
plt.gca().set_yticks(range(-10,15,5))
plt.gca().yaxis.grid(which = 'major', color = '#e1e1e1', linestyle = '--', linewidth = 0.5)

plt.legend(title='Country')

ylabs = plt.gca().get_yticks().astype(int).astype(str)
ylabs  = [str(x) + '%' for x in ylabs]
plt.gca().set_yticklabels(ylabs)
```
